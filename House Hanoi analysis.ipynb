```python
# Transform data in the 3 columns 'square_meter', 'bedroom_no', 'price' so that it can be converted to the correct types.
properties['square_meter'] = properties['square_meter'].apply(lambda x: int(x[:4]))
properties['bedroom_no'] = properties['bedroom_no'].apply(lambda z: z.split()[0])
properties['price'] = properties['price'].apply(lambda y: y.split()[0]+ ' ' + y.split()[1])
```

```python
properties['bedroom_no'] = properties['bedroom_no'].astype('int32')
```
